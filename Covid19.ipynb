```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('covid_data.csv')
df
```

```python
%matplotlib inline
plt.figure(figsize=(50, 16))
ax = plt.axes()
ax.grid(linewidth=.4)
ax.set_facecolor('black')
ax.set_xlabel(df["Date"], size=25, color='skyblue')
ax.set_ylabel(df["Daily Deceased"], size=25, color='skyblue')

plt.xticks(rotation='vertical',size='20',color='black')
plt.yticks(size=20,color='black')
plt.tick_params(size=20,color='black')
plt.title("COVID-19 IN : Daily Confirmed\n",
          size=50,color='#28a9ff')

for i,j in zip(df['Date'],df['Daily Deceased']):
    ax.annotate(str(j),xy=(i,j+1),
                color='yellow',size='20')

ax.annotate('Second Lockdown 15th April',
            xy=(76, 25),
            xytext=(70,10),
            color='white',
            size='25',
            arrowprops=dict(color='white',
                            linewidth=0.025))

ax.plot(df['Date'], df['Daily Deceased'], color='skyblue', marker='o', linewidth=4, markersize=15, markeredgecolor='blue')
```
